# Retrieve RSS Feed Content for Given Feed List

## Load useful libraries

```python
import os
import pickle
import json
```

```python
from strategic_reports.daily.inators.content_fetch_inator import fetch_rss_content, qa_fetch_rss_content
```

## User settings

```python
filename_json_feeds_list = os.environ['STRATEGIC_REPORTS_HOME'] + '/data/daily/feeds_ai.json'
filename_output_pickled = os.environ['STRATEGIC_REPORTS_HOME'] + '/output/daily/feed_content.pickled'
filename_output_exceptions = os.environ['STRATEGIC_REPORTS_HOME'] + '/output/daily/exceptions_feed_pull.json'

hours_cutoff = 24
number_of_feed_retrieval_processes = 15
```

## Fetch the RSS content for the feed

```python
list_feeds = fetch_rss_content(
    filename_json_feeds_list,
    number_of_feed_retrieval_processes = number_of_feed_retrieval_processes,
    hours_cutoff = hours_cutoff,
)
```

## QA

```python
qa_success = qa_fetch_rss_content(list_feeds)
```

## Save results

```python
if qa_success:
    with open(filename_output_pickled, 'wb') as f:
        pickle.dump(list_feeds, f)
```

## Highlight exceptions

```python
def find_cases_with_exceptions(feed_pull_results):
    list_exception_cases = []
    for item in feed_pull_results:
        if 'exception' in item:
            to_keep = {
                'title' : item['title'],
                'url' : item['url'],
                'exception' : item['exception'],
            }
            list_exception_cases.append(to_keep)
    return list_exception_cases
```

```python
list_exceptions = find_cases_with_exceptions(list_feeds)
```

```python
with open(filename_output_exceptions, 'w') as f:
    f.write(json.dumps({'exceptions' : list_exceptions}, indent = 4) + '\n')
```
