# Generate Strategic Summary Using an LLM

## Load useful libraries

```python
import os
```

```python
from tools_and_shortcuts.ai.llm.OllamaInator import OllamaInator
```

```python
from strategic_reports.daily.prompts.prompts import prompt_strategic_summary
```

## User settings

```python
model_to_use_for_strategic_summary = 'gpt-oss:120b'
filename_markdown_report_summaries_and_tags = os.environ['STRATEGIC_REPORTS_HOME'] + '/output/daily/report_summaries_and_tags.md'
filename_markdown_report_strategy = os.environ['STRATEGIC_REPORTS_HOME'] + '/output/daily/report_strategy.md'
```

## Initialize Ollama

```python
oi = OllamaInator()
```

## Load the summaries and tags report

```python
with open(filename_markdown_report_summaries_and_tags) as f:
    report_summaries = '\n'.join([x.strip() for x in f.readlines()])
```

## Run the strategic summarization process using an LLM

```python
response = oi.run_generate(
    prompt_strategic_summary + report_summaries,
    model = model_to_use_for_strategic_summary,
)
report_strategy = response['response_text']
```

## Save results in Markdown format

```python
with open(filename_markdown_report_strategy, 'w') as f:
    f.write(report_strategy + '\n')
```
