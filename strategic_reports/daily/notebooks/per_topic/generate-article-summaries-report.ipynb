# Generate Per-Article Summaries and Tags Report

## Load useful libraries

```python
import os
import json
```

```python
from strategic_reports.daily.inators.report_inator import produce_article_summary_markdown_report
```

## User settings

```python
filename_json_summaries_and_tags = os.environ['STRATEGIC_REPORTS_HOME'] + '/output/daily/summaries_and_tags.json'
filename_markdown_report_summaries_and_tags = os.environ['STRATEGIC_REPORTS_HOME'] + '/output/daily/report_summaries_and_tags.md'
```

## Load the summaries and tags JSON produced previously by an LLM

```python
with open(filename_json_summaries_and_tags) as f:
    dict_summaries_tags = json.load(f)
```

## Generate summaries portion of the report

```python
report_summaries = produce_article_summary_markdown_report(dict_summaries_tags, preamble = False)
```

## Save summaries report

```python
with open(filename_markdown_report_summaries_and_tags, 'w') as f:
    f.write(report_summaries)
```
